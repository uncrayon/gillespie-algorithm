# A _While_ Test

So, the great problem of our system is that time advance a Poisson process (somthing exponential) and it is time-expensive. A way to solve this problem is to try to optimize first our process

First, let us define two random exp functions

```python
def random_exp(a0):
    r = np.random.random()
    #tau = (1/a0)*np.log(1/r)
    return (1/a0)*np.log(1/r)
```

```python
def random_exp_np(a0):
    return np.random.exponential(1/a0)
```

```python
%timeit -r 15 random_exp(2)
```

```python
%timeit -r 15 random_exp_np(2)
```

The times are pretty close but the harcoded implementation got a lower std. dev. which means is more stable and predictable. 

Now, lets njit the functions

```python
random_exp_jitted = njit(random_exp)
random_exp_np_jitted = njit(random_exp_np)
```

And to end let us run the timeit test again.

```python
direct_method = %timeit -o -r 15 random_exp_jitted(2.0)
```

```python
numpy_method = %timeit -o -r 15 random_exp_np_jitted(2.0)
```

```python
print('Worst in direct was {:.2} and numpy as {:.2}'.format(direct_method.worst,numpy_method.worst))
```

```python
direct_method.worst<numpy_method.worst
```

Even in the worst was faster our direct method. 

Now, let us notice that in our direct method we depend on a numpy medthod: `numpy.log`. Let us now see if our direct method gets faster with `math.log` method

```python
def random_exp_numpy(a0):
    r = np.random.random()
    #tau = (1/a0)*np.log(1/r)
    return (1/a0)*np.log(1/r)

def random_exp_math(a0):
    r = np.random.random()
    #tau = (1/a0)*np.log(1/r)
    return (1/a0)*math.log(1/r)
```

```python
numpy_log = %timeit -o -r 15 random_exp_numpy(2.0)
```

```python
math_log = %timeit -o -r 15 random_exp_math(2.0)
```

```python
print('Worst in numpy was {:.2} and math as {:.2}'.format(numpy_log.worst,math_log.worst))
```

```python
print('Is math best in the direct method than numpy? {}'.format(numpy_log.worst>math_log.worst) )
```

Now let us njit our functions

```python
random_exp_numpy_jitted = njit(random_exp_numpy)
random_exp_math_jitted = njit(random_exp_math)
```

```python
numpy_log_jitted = %timeit -o -r 15 random_exp_numpy_jitted(2.0)
```

```python
math_log_jitted = %timeit -o -r 15 random_exp_math_jitted(2.0)
```

```python
print('Worst in numpy jitted was {:.2} and math jitted as {:.2}'.format(numpy_log_jitted.worst,math_log_jitted.worst))
```

```python
print('Remains math best in the direct method than numpy once jitted? {}'.format(numpy_log_jitted.worst>math_log_jitted.worst) )
```

```python
numpy_log_jitted.worst<math_log_jitted.worst
```

```python
numpy_log_jitted.best<math_log_jitted.best
```

```python
numpy_log_jitted.average<math_log_jitted.average
```

Then the direct method is faster with njit method. 

Now lets do a pair of while loops:

```python
@njit
def random_exp_numpy(a0):
    """
    This functions generates a exponential random 
    variable with mean and std. dev. 1/a0
    """
    r = np.random.random()
    return (1/a0)*np.log(1/r)
```

```python
def Loop_1(t0=0.0,t_max=10.0,a0=2.0):
    while t0<t_max:
        t0 += random_exp_numpy(a0)
        
def Loop_2(t0=0.0,t_max=10.0,a0=2.0):
    while True:
        if t_max<t0:
            break
        t0 += random_exp_numpy(a0)
```

```python
%timeit -r 15 Loop_1()
```

```python
%timeit -r 15 Loop_2()
```

Now, let us consider a greater time, like `t_max = 100`.

```python
%timeit -r 15 Loop_1(t_max=100.0)
```

```python
%timeit -r 15 Loop_2(t_max=100.0)
```

Keeping the function with stop condition on `while` gets best performance. Now, let us try to profiling the functions

```python
%lprun -f Loop_1 Loop_1(t_max=100.0)
```

```python
%lprun -f Loop_2 Loop_2(t_max=100.0)
```

Keeping the condition on `while` give is a best performance cause we stop without executing an external `break`, thats why we save time. 

Now, let us see if there is performance _njiting_ the functions.

```python
Loop_1_jitted = njit(Loop_1)
Loop_2_jitted = njit(Loop_2)
```

```python
%timeit -r 15 Loop_1_jitted(t_max=100.0)
```

```python
%timeit -r 15 Loop_2_jitted(t_max=100.0)
```

It improved for almost a factor of 2x! (~1.8x)

Now, let us njit our functions

```python
def append_list(a,b):
    a.append(b)
    return a
#    return a + [b]
    
def append_numpy(a,b):
    return np.append(a,b.reshape(1,len(b)),axis=0)

def stack_numpy(a,b):
    s = a.shape[1]
    return np.vstack((a,b.reshape(1,s)))
```

```python
%timeit -r 7 append_numpy_jitted(a_numpy,dummy_array)
```

But as a whole we can use append!

```python
def my_func_list(t_max):
    a = [[1,2,3]]
    t=0
    while t<t_max:
        a.append([2,3,4])
        t += 1
    return a

def my_func_append(t_max):
    a = np.array([[1,2,3]])
    t=0
    while t<t_max:
        a = np.append(a,np.array([2,3,4]).reshape(1,3),axis=0)
        t += 1
    return a

def my_func_vstack(t_max):
    a = np.array([[1,2,3]])
    t=0
    while t<t_max:
        a = np.vstack((a,np.array([2,3,4]).reshape(1,3)))
        t += 1
    return a
```

```python
my_func_jitted = njit(my_func_list)
my_func_append_jitted = njit(my_func_append)
my_func_vstack_jitted = njit(my_func_vstack)
```

```python
%timeit -r 7 my_func(100)
```

```python
%timeit -r 7 my_func_jitted(100)
```

```python
%timeit -r 7 my_func_append(100)
```

```python
%timeit -r 7 my_func_append_jitted(100)
```

```python
%timeit -r 7 my_func_vstack(100)
```

```python
%timeit -r 7 my_func_vstack_jitted(100)
```

# An experimental application of SSA

The last form of [SSA that we develop](./Epoch_1.ipynb#1.-An-Introduction-to-Stochastic-Simulation-of-Chemical-Kinetics) is able to get a vector of initial poulations, an initited time vector, a function, a state-change vector and a max time, lets rebuild it!

```python
@njit
def random_exp_numpy(a0):
    """
    This functions generates a exponential random 
    variable with mean and std. dev. 1/a0
    """
    r = np.random.random()
    return (1/a0)*np.log(1/r)

@njit
def sum_vectors_list(a,b):
    l = []
    n = len(a)
    for i in prange(n):
        l.append(a[i]+b[i])
    return l

@njit
def sum_vect(a):
    n = len(a)
    k = 0.0
    for i in prange(n):
        k += a[i]
    return k

@njit
def div_vect(a,a0):
    n = len(a)
    b = []
    for i in prange(n):
        b.append(a[i]/a0)
    return b
```

```python
_=random_exp_numpy(2)
_=sum_vectors_list([1,2],[1,2])
_=sum_vect([1,2])
_=div_vect([1,2],1)
```

```python
def SSA_v0(x,t,a,v,t_max,seed = 42):
    """
    A toy version of the SSA described in the notebook
    
    Params
    ---------
    x : array_like
        A vector of states vectors, if len(x)>1 then is going to be used x[-1]
    t : array_like
        A vecttor of timesteps, t0 = t[-1]
    a : func
        The vectorial function form of all propensity functions
    v : array_like
        The state change vector or matrix
    t_max : float
        The condition time to stop
    seed : int
        Seed to secure reproducibility
        
    Returns
    ----------
    x : array_like
        x.shape = (m,n) where m is number if reactions and n the number of
        molecular species
        The whole system vector-population after a R_{i} in a time t_{i}
    t : array_like
        The time steps at which each reaction occurs
    """
    np.random.seed(seed)
    idx = np.arange(len(v))
    t0,tau=0.0,0.0
    while t0+tau<t_max:
        x0,t0 = x[-1],t[-1]
        
        a_vector = a(x0)

        a_0 = np.sum(a_vector)
            
        # Tau dist prob of generating function is a_{0}*exp(-tau*a_{0}) therefore the following is more stable than the later tau-generator
        tau = np.random.exponential(scale=1/a_0)
        # the dist prob of generating function of j is a_{j}/a_{0}, a point probability so the following is more stable than the later j-generator
        j = random.choices(idx,weights=a_vector/a_0)[0]
                
        t = np.append(t,t0+tau)
        x = np.append(x,[x0+v[j]],axis=0)
        #print(t0+tau,a_vector,a_0,j)
        
    return x,t
```

```python
def SSA_v1(x0,t0,a,v,t_max,seed = 42):
    """
    A toy version of the SSA described in the notebook
    
    Params
    ---------
    x : array_like
        A vector of states vectors, if len(x)>1 then is going to be used x[-1]
    t : array_like
        A vecttor of timesteps, t0 = t[-1]
    a : func
        The vectorial function form of all propensity functions
    v : array_like
        The state change vector or matrix
    t_max : float
        The condition time to stop
    seed : int
        Seed to secure reproducibility
        
    Returns
    ----------
    x : array_like
        x.shape = (m,n) where m is number if reactions and n the number of
        molecular species
        The whole system vector-population after a R_{i} in a time t_{i}
    t : array_like
        The time steps at which each reaction occurs
    """
    np.random.seed(seed)
        
    idx = np.arange(len(v))
    T = [t0]
    t0 = T[0]
    X = [[i for i in x0]]
    x0 = X[0]
    n = len(v)
    while t0<t_max:
        
        a_vector = a(x0)
        #a_0 = sum_vect(a_vector)
        a_0 = sum(a_vector)
        
        # Tau dist prob of generating function is a_{0}*exp(-tau*a_{0}) therefore the following is more stable than the later tau-generator
        tau = random_exp_numpy(a_0)
        # the dist prob of generating function of j is a_{j}/a_{0}, a point probability so the following is more stable than the later j-generator
        j = random.choices(idx,weights=map(truediv,a_vector,[a_0]*n))[0]
                
        T.append(t0+tau)
        X.append( list(map(add,x0,v[j])) )
        
        s = len(T)-1
        t0 = T[s]
        x0 = X[s]
                                   
    return X,T
```

Now let us use it in the [brusselator](./Epoch_1.ipynb#2.4-The-Brusselator-(cfr.-[2]))

```python
@njit
def R_j(Y,k1=5000.0,k2=50.0,c3=0.00005,c4=5.0):
    return [k1,k2*Y[0],c3*Y[1]*Y[0]*0.5*(Y[0]-1),c4*Y[0]]

def R(Y,k1=5000.0,k2=50.0,c3=0.00005,c4=5.0):
    return [k1,k2*Y[0],c3*Y[1]*Y[0]*0.5*(Y[0]-1),c4*Y[0]]
```

```python
_=R_j([10,10])
```

```python
v = [[1,0],
    [-1,1],
    [1,-1],
    [-1,0]]
```

```python
%timeit -r 1 SSA_v1(x0=[1000,2000],t0=0,a=R,v=v,t_max=14)
```

```python
%lprun -f SSA_v1 SSA_v1(x0=[1000,2000],t0=0,a=R,v=v,t_max=14)
```

```python
%lprun -f SSA_v0 SSA_v0(x=np.asarray([[1000,2000]]),t=[0],a=R,v=v,t_max=2)
```

With this optimization sessions we got finished a simulation for `t_max = 14` in 22.5s while the later presented function for a `t_max = 2` got 31s. As the differences in time are exponential-distributed and the ocurrence of reactions are poisson processes then we got an exponential-order optimization.

```python
X,T = SSA_v1(x0=[1000,2000],t0=0,a=R,v=v,t_max=14)
```

```python
x,t = np.asarray(X),np.asarray(T)

fig,ax = plt.subplots(1,1,dpi=120,figsize=(15,7))

ax.plot(t,x[:,0],label=r'$Y_{1}$ ')
ax.plot(t,x[:,1],label=r'$Y_{2}$')

ax.set_xlabel(r'$t$')
ax.set_ylabel(r'Molecular Populations $Y_{1}(t)$, $Y_{2}(t)$')
ax.set_title(r'Evolution of $X$, $Y_{1}$ and $Y_{2}$ for Lotka Reactions')

ax.legend()

plt.show()
```
